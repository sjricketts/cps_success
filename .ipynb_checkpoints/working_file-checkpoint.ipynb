```python
##################read in teacher salary doc as csv
path = ("resources/EmployeePositionRoster_12312019.csv")
salary_roster = pd.read_csv(path)
```

```python
#create new dataframe with only relevant columns
salary_df = salary_roster[["Department", "Annual Salary", "Job Title"]]
```

```python
#Rename columns
Renamed_Salary = salary_df.rename(columns={"Department": "department", "Annual Salary": "Annual_Salary", "Job Title": "Job_Title"})
```

```python
#Make a new DF with only teachers in the list
teachers_salary = Renamed_Salary[Renamed_Salary.Job_Title.str.contains('Teacher',case=False)]
```

```python
#group by school to find the mean
mean_salary = teachers_salary.groupby(["department"]).mean()
```
